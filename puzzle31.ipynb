## Create dataframes per lamp which is on

```python
lamp1 = puzzle31[puzzle31['lamp1']==1].copy()
lamp1.drop(columns=['lamp1', 'lamp2', 'lamp3', 'lamp4'], inplace=True)

lamp2 = puzzle31[puzzle31['lamp2']==1].copy()
lamp2.drop(columns=['lamp1', 'lamp2', 'lamp3', 'lamp4'], inplace=True)

lamp3 = puzzle31[puzzle31['lamp3']==1].copy()
lamp3.drop(columns=['lamp1', 'lamp2', 'lamp3', 'lamp4'], inplace=True)

lamp4 = puzzle31[puzzle31['lamp4']==1].copy()
lamp4.drop(columns=['lamp1', 'lamp2', 'lamp3', 'lamp4'], inplace=True)

print(f"lamp1: {lamp1.shape}")
print(f"lamp2: {lamp2.shape}")
print(f"lamp3: {lamp3.shape}")
print(f"lamp4: {lamp4.shape}")
```

## Create dataframes & heatmaps where one lamp is on and other lamps are off

### lamp 1 - all on or all off

```python
lamp1_only = puzzle31[(puzzle31['lamp1']==1) & (puzzle31['lamp2']==0) & (puzzle31['lamp3']==0) & (puzzle31['lamp4']==0)].copy()
lamp1_only.drop(columns=['lamp1', 'lamp2', 'lamp3', 'lamp4'], inplace=True)
lamp1_only_ax = sns.heatmap([[lamp1_only['A1'].sum(), lamp1_only['A2'].sum(), lamp1_only['A3'].sum()], [lamp1_only['B1'].sum(), lamp1_only['B2'].sum(), lamp1_only['B3'].sum()], [lamp1_only['C1'].sum(), lamp1_only['C2'].sum(), lamp1_only['C3'].sum()]], linewidths=0.5, cmap="Blues", center=0)
```

### lamp 2 - diagonals

```python
lamp2_only = puzzle31[(puzzle31['lamp1']==0) & (puzzle31['lamp2']==1) & (puzzle31['lamp3']==0) & (puzzle31['lamp4']==0)].copy()
lamp2_only.drop(columns=['lamp1', 'lamp2', 'lamp3', 'lamp4'], inplace=True)
lamp2_only_ax = sns.heatmap([[lamp2_only['A1'].sum(), lamp2_only['A2'].sum(), lamp2_only['A3'].sum()], [lamp2_only['B1'].sum(), lamp2_only['B2'].sum(), lamp2_only['B3'].sum()], [lamp2_only['C1'].sum(), lamp2_only['C2'].sum(), lamp2_only['C3'].sum()]], linewidths=0.5, cmap="Blues", center=0)
```

### lamp 3 - verticals

```python
lamp3_only = puzzle31[(puzzle31['lamp1']==0) & (puzzle31['lamp2']==0) & (puzzle31['lamp3']==1) & (puzzle31['lamp4']==0)].copy()
lamp3_only.drop(columns=['lamp1', 'lamp2', 'lamp3', 'lamp4'], inplace=True)
lamp3_only_ax = sns.heatmap([[lamp3_only['A1'].sum(), lamp3_only['A2'].sum(), lamp3_only['A3'].sum()], [lamp3_only['B1'].sum(), lamp3_only['B2'].sum(), lamp3_only['B3'].sum()], [lamp3_only['C1'].sum(), lamp3_only['C2'].sum(), lamp3_only['C3'].sum()]], linewidths=0.5, cmap="Blues", center=0)
```

```python
lamp3_0 = sns.heatmap([lamp3_only.iloc[0][0:3],lamp3_only.iloc[0][3:6], lamp3_only.iloc[0][6:9]], linewidths=0.5, cmap="Blues", center=0)
```

### lamp 4 - horizontals

```python
lamp4_only = puzzle31[(puzzle31['lamp1']==0) & (puzzle31['lamp2']==0) & (puzzle31['lamp3']==0) & (puzzle31['lamp4']==1)].copy()
lamp4_only.drop(columns=['lamp1', 'lamp2', 'lamp3', 'lamp4'], inplace=True)
lamp4_only_ax = sns.heatmap([[lamp4_only['A1'].sum(), lamp4_only['A2'].sum(), lamp4_only['A3'].sum()], [lamp4_only['B1'].sum(), lamp4_only['B2'].sum(), lamp4_only['B3'].sum()], [lamp4_only['C1'].sum(), lamp4_only['C2'].sum(), lamp4_only['C3'].sum()]], linewidths=0.5, cmap="Blues", center=0)
```

```python
lamp4_0 = sns.heatmap([lamp4_only.iloc[0][0:3],lamp4_only.iloc[0][3:6], lamp4_only.iloc[0][6:9]], linewidths=0.5, cmap="Blues", center=0)
```

## size of groupby

### lamp 4

```python
lamp4_only.groupby(['A1', 'A2', 'A3']).size()
```

```python
lamp4_only.groupby(['A1', 'B1', 'C1']).size()
```

There are no input combinations where A1, B1, and C1 are all of or all on.

More similar to all 1s than to all As.
